## Finding Average Price for  Top Brands

Now that we have a good idea of what brands are available, I'm going to determine the mean price of the top brands. In the cell above, I created a list called 'top_brands'which shows the top 10 brands by number of listings.  Next, I am going to find the mean price of the listed brands and create a dictionary to display the mean prices in descending order.

```python
import operator
top_cars_mean = {}

for cars in top_brands:
    car_brand = autos[autos['brand']==cars]
    mean_price = car_brand['price'].mean().round(2)
    top_cars_mean[cars] = mean_price
    
by_price = sorted(top_cars_mean.items(), key=operator.itemgetter(1), reverse=True)
by_price
```

From our top brands, we can start to distinguish groups of cars based on mean price. It's clear that Audi, BMW, and Mercedes Benz are in a class of their own. Ford, SEAT, Opel, and the rest near the bottom round out the second group. In between is Volkswagen, not really in either group. 

Being a car guy, one thing I noticed is the distinct leaning for German cars. Audi/Volkswagen, BMW, and Mercedes Benz are cleary separate from the rest. Opel is technically a German brand, but is owned by General Motors and typically produces cars on shared platforms with Chevy, Buick, and Cadillac, etc. Despite being relatively similar to Volkswagen in terms of sticker price, the resale values are much lower for American operated Opel and German operated Volkswagen.

With this fact in mind, I want to investigate the differences between the cars for sale to see if this price difference is indeed a favorability for german cars, or if it's just a coincedence. To do that, I will take a look at the average mileage of the brands.

```python
mean_km = {}

for cars in top_brands:
    car_brand = autos[autos['brand']==cars]
    odometer_mean = car_brand['odometer_km'].mean().round()
    mean_km[cars] = odometer_mean
    
mean_km
```

```python
brand_mean_km = pd.Series(mean_km)
brand_mean_price = pd.Series(top_cars_mean)
bmk_df = pd.DataFrame(brand_mean_km, columns=['mean_km'])
bmp_df = pd.DataFrame(brand_mean_price, columns=['mean_price'])
bmk_df.insert(1, 'mean_price', brand_mean_price)

bmk_df
```
